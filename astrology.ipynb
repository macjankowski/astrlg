```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
import pandas as pd
```

```python
mu = 31+28+31+15
variance = 10
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```

```python
sample_count = 1000000
```

```python
s = np.random.normal(mu, sigma, sample_count)
```

```python
min(s), max(s)
```

```python
counts, edges, plot = plt.hist(s, bins=list(range(0,365)))
```

```python
counts = np.append(counts, [0])
```

```python
len(counts)
```

```python
probs = pd.DataFrame({
    "day": edges+1,
    "count": counts/sample_count
})
```
